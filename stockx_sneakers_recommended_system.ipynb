```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
data['tags'] = data['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=3000,stop_words='english')
```

```python
cv.fit_transform(data['tags']).toarray()
```

```python
cv.fit_transform(data['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(shoe):
    shoe_index = data[data['product_name'] == shoe].index[0]
    distance = similarity[shoe_index]
    shoes_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in shoes_list:
        print(data.iloc[i[0]].product_img_url)
        print(data.iloc[i[0]].product_name)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
recommend('Nike Dunk Low Retro White Black Panda (2021)')
```
